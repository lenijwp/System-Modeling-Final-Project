```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import torch
import torch_cluster
from typing import Optional
import warnings
from sklearn import metrics
warnings.filterwarnings("ignore")
otc=pd.read_csv('bits-otc.csv')
alp=pd.read_csv('bits-alp.csv')
device=torch.device('cuda:0')
alp
```

```python

def get_edge_tuple( raw:pd.DataFrame):
    tuple_list=[]
    for source,target,rating in zip(raw['SOURCE'],raw['TARGET'],raw['RATING']):
        tuple_list.append((source,target,rating))
    return tuple_list

def draw_raw_graph(raw:pd.DataFrame,range_of_data=10000,color=None):
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)[:range_of_data]
    G.add_weighted_edges_from(node_list)
    nx.draw(G)
    plt.show()
```

```python
#alp shape
draw_raw_graph(alp)
```

```python
#otc shape
draw_raw_graph(otc)
```

```python
def get_feature_matrix(A):
    A = np.array(A)
    I = np.eye(A.shape[0])
    A_hat = A + I
    D_hat = int (np.array(np.sum(A_hat, axis=0))[0])
    H_2 = A_hat * D_hat

    return H_2
print(get_feature_matrix([[1,1],[1,2]]))
```

```python
from sklearn.cluster import KMeans
def graclus_cluster(raw:pd.DataFrame):
    def _graclus (st,ed,weight):
        num_nodes = max(int(st.max()), int(ed.max())) + 1
        mask = st != ed
        st, ed = st[mask], ed[mask]
        weight = weight[mask]
        # Randomly shuffle
        perm = torch.argsort(st)
        st, ed = st[perm], ed[perm]
        if weight is not None:
            weight = weight[perm]

        deg = st.new_zeros(num_nodes)
        deg.scatter_add_(0, st, torch.ones_like(st))
        ptr = st.new_zeros(num_nodes + 1)
        torch.cumsum(deg, 0, out=ptr[1:])

        return torch.ops.torch_cluster.graclus(ptr, ed, weight)
    source = torch.tensor(np.array(raw['SOURCE']))
    target = torch.tensor (np.array(raw['TARGET']))
    weight = torch.tensor(np.array(raw['RATING']))
    cluster = _graclus(source, target, weight)
    return cluster

def fps_cluster(raw:pd.DataFrame):
    def fps(src: torch.Tensor, batch, ratio, random_start=True):  
        r = torch.tensor(ratio, dtype=src.dtype, device=src.device)
        batch_size = int(batch.max()) + 1

        deg = src.new_zeros(batch_size, dtype=torch.long)
        deg.scatter_add_(0, batch, torch.ones_like(batch))

        ptr = deg.new_zeros(batch_size + 1)
        torch.cumsum(deg, 0, out=ptr[1:])
        return torch.ops.torch_cluster.fps(src, ptr, r, random_start)
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)
    G.add_weighted_edges_from(node_list)
    A=np.array(nx.adjacency_matrix(G).todense())
    A=get_feature_matrix(A)
    x = torch.tensor(A)
    batch = torch.tensor([0]*x.shape[0])
    cluster = fps(x, batch, ratio=0.5, random_start=False)
    return cluster
    
def spectral_cluster(raw:pd.DataFrame,k=1000):
    def compute_affinity_matrix(X):
        l2_norms = np.linalg.norm(X, axis=1)
        X_normalized = X / l2_norms[:, None]
        cosine_similarities = np.matmul(X_normalized, np.transpose(X_normalized))
        affinity = (cosine_similarities + 1.0) / 2.0
        return affinity
    def compute_sorted_eigenvectors(A):
        eigenvalues, eigenvectors = np.linalg.eig(A)
        eigenvalues = eigenvalues.real
        eigenvectors = eigenvectors.real
        index_array = np.argsort(-eigenvalues)
        w = eigenvalues[index_array]
        v = eigenvectors[:, index_array]
        return w, v
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)
    G.add_weighted_edges_from(node_list)
    A=np.array(nx.adjacency_matrix(G).todense())
    A=get_feature_matrix(A)
    X = torch.tensor(A)

    affinity = compute_affinity_matrix(X)
    (eigenvalues, eigenvectors) = compute_sorted_eigenvectors(
        affinity)

    spectral_embeddings = eigenvectors[:, :k]
    kmeans_clusterer = KMeans(
        n_clusters=k,
        init="k-means++",
        max_iter=300,
        random_state=0)
    cluster = kmeans_clusterer.fit_predict(spectral_embeddings)
    return cluster
    
def grid_cluster(raw:pd.DataFrame):
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)
    G.add_weighted_edges_from(node_list)
    A=np.array(nx.adjacency_matrix(G).todense())
    A=get_feature_matrix(A)
    X = torch.tensor(A)
    cluster =torch.ops.torch_cluster.grid(X,X.max(1)[0] ,None,None)
    cluster = np.array(cluster).tolist()
    dic={}
    le=0
    for i in cluster:
        if(i not in dic):
            dic[i]=le
            le+=1
    cluster=[dic[i] for i in cluster]
    return cluster

def KNN_cluster(raw:pd.DataFrame,k=1000):
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)
    G.add_weighted_edges_from(node_list)
    A=np.array(nx.adjacency_matrix(G).todense())
    A=get_feature_matrix(A)
    X = torch.tensor(A)
    kmeans_clusterer = KMeans(
        n_clusters=k,
        init="k-means++",
        max_iter=300,
        random_state=0)
    cluster = kmeans_clusterer.fit_predict(X)
    return cluster
```

```python
def draw_clustered_graphy(cluster,raw):
    G = nx.DiGraph()
    node_list=get_edge_tuple(raw)
    G.add_weighted_edges_from(node_list)
    cluster=np.array(cluster).tolist()
    all_dic={}
    pt=0
    for i in range(len(cluster)):
        if (cluster[i] not in all_dic):
            all_dic[cluster[i]]=pt
            pt+=1
    for i in range(len(cluster)):
        cluster[i]=all_dic[cluster[i]]
    max_l=max(cluster)+1
    cluster = [i/max_l for i in cluster ][:len(G.nodes)]
    if(len(G.nodes)>len(cluster)): cluster += [0]*(len(G.nodes)-len(cluster))
    nx.draw(G, node_color=cluster)
    plt.show()
```

```python
# graclus
a=graclus_cluster(alp)
b=graclus_cluster(otc)
draw_clustered_graphy(a,alp)
```
